```python
def check_position(table, row, col):
    if database_split.loc[row, col] == 'Cs' or database_split.loc[row, col] == ' Cs':
        temp = pd.DataFrame({'Cs': float(dataratio_split.loc[row, col]), "MA":[pd.NA], "FA":[pd.NA]}, index=[0])
        temp = pd.concat([table, temp], ignore_index= True)
        return temp
    elif database_split.loc[row, col] == 'MA' or database_split.loc[row, col] == ' MA':
        temp = pd.DataFrame({'MA': float(dataratio_split.loc[row, col]), "Cs":[pd.NA], "FA":[pd.NA]}, index=[0])
        temp = pd.concat([table, temp], ignore_index= True)
        return temp
    elif database_split.loc[row, col] == 'FA' or database_split.loc[row, col] == ' FA':
        temp = pd.DataFrame({'FA': float(dataratio_split.loc[row, col]), "MA":[pd.NA], "Cs":[pd.NA]}, index=[0])
        temp = pd.concat([table, temp], ignore_index= True)
        return temp
    else:
        return None
```

```python
#The only ones needed are: Cs MA FA Pb Cl I Sn Br so for a ions, get "Cs MA FA"
table = pd.DataFrame(columns=['Cs', 'MA', 'FA'])

for y in range(0,4):
    for x in range(0,43):  #43378
        table = check_position(table, x, y)

table
```
