```python
import numpy as np
import pandas as pd

database = pd.read_csv("Data/Perovskite_database_content_all_data.csv", low_memory=False)
```

```python
threshold = 0.7

# Calculate the percentage of missing values for each column
missing_percentage = database.isna().mean()

# Filter columns based on the threshold
filtered_df = database.loc[:, missing_percentage <= threshold]

# Output the result
print("Original DataFrame:")
print(database)
print("\nFiltered DataFrame (columns with less than or equal to 50% missing values):")
filtered_df
#column types: 
#ETL, HTL, Perovskite, JV, Outdoor, EQE, Stability, Stablised performance, Encapsulation, Add-layer-front, 
#Add-layer-back, Backcontact, Substrate, Module, Cell
```
